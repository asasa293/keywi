## Conseguimos el ponderado de las ventas

```python
sales = pd.read_excel(r"C:\Users\miste\Downloads\Sales_auto.xlsx", usecols="B:N", skiprows=4)
sales = sales.drop(index=sales.index[-4:])
sales
```

```python
sales['Month'] = sales['Date'].dt.month
sales['Year'] = sales['Date'].dt.year
```

```python
sales
```

```python
grouped_data = sales.groupby(['Month', 'Class'])['Amount'].sum()

# reset index to make the grouped columns into regular columns
grouped_data = grouped_data.reset_index()

# show the resulting dataframe
grouped_data
```

```python
grouped_data = grouped_data.drop(index=grouped_data[grouped_data['Class']
                                                    .isin(['Garland LLC:JCH Organic SAC',
                                                           'Garland LLC:S.A.T. Montealbo',
                                                           'Garland LLC:Natural Produce of Peru',
                                                           'El Sol Brands:JCH Organic SAC',
                                                           'Eshkol Premium SAS',
                                                           'HLB Specialties LLC:JCH Organic SAC',
                                                           'Plantar de Colombia SAS',
                                                           'Smart Group Traders:Natural Produce of Peru'
                                                          ]
                                                         )
                                                   ]
                                 .index)
```

```python
grouped_data
```

```python
# group the dataframe by 'Month'
grouped = grouped_data.groupby('Month')

# create an empty dataframe to store the results
ponderado = pd.DataFrame()

# loop through each group
for month, group in grouped:
    # calculate the total amount for this month
    total = group['Amount'].sum()
    
    # calculate the percentage for each class
    group['%'] = group['Amount'] / total * 100
    
    # add the group to the result dataframe
    ponderado = ponderado.append(group)

# sort the result dataframe by 'Month' and 'Class'
ponderado = ponderado.sort_values(['Month', 'Class'])

# reset the index
ponderado = ponderado.reset_index(drop=True)

ponderado = ponderado.drop(ponderado[ponderado['%'] == 0].index)
```

## Conseguimos los COGS para identificar cuales no tienen Class

```python
df = pd.read_excel(r"C:\Users\miste\Downloads\COGS.xlsx", usecols="B:R", skiprows=4)
```

```python
df
```

```python
df = df.drop(index=df.index[-4:])
df.tail(6)
```

```python
df[['AWB', 'HAWB', 'Lot']] = df['Memo/Description'].str.split(',', expand=True)
df[['Category', 'Subcategory', 'SKU']] = df['Product/Service'].str.split(':', expand=True)
df['SKU'] = df['SKU'].str.replace('Peeled ', '').str.replace('Organic ', '')
df['SKU'] = df['SKU'].str.replace('Conventional ', '').str.replace('Organic ', '')
df['SKU'] = df['SKU'].str.replace(' 4.0 Oz', '').str.replace(' 2.5 Oz', '')
df['SKU'] = df['SKU'].str.replace(' 0.75 Oz', '').str.replace(' 8.0 Oz', '')
df['SKU'] = df['SKU'].str.replace(' 3.0 Oz', '').str.replace(' 1.25 Oz', '')
df['SKU'] = df['SKU'].str.replace(' 1.0 Oz', '')
```

```python
def clean_product(product):
    if isinstance(product, str) and '(' in product:
        cleaned = product[:product.index('(')].strip()
    else:
        cleaned = product
    return cleaned

# apply the function to the Product column
df['SKU'] = df['SKU'].apply(clean_product)
```

```python
df.head()
```

```python
# df.to_excel(r"C:\Users\miste\Downloads\COGS_Guias.xlsx")
```

```python
empty_class = df[df['Class'].isna()]

# Delete rows with empty "Class" from original dataframe
df = df.dropna(subset=['Class'])
```

```python
empty_class
```

```python
empty_class['Amount'].sum()
```

```python
# group the empty_class DataFrame by the "Date" column and calculate the sum of the "Amount" column for each group
empty_class_summed = empty_class.groupby('Date').agg({'Amount': 'sum'})

# create a new DataFrame to store the summarized information
new_df = pd.DataFrame(columns=empty_class.columns)

# loop through each group in the empty_class_summed DataFrame
for date, group in empty_class_summed.groupby('Date'):
    # create a dictionary for the new row
    new_row_dict = {
        'Date': date,
        'Transaction Type': '',
        'Num': '',
        'Name': '',
        'Memo/Description': 'Shipping/Handling/Vacuum/Import Fee/etc',
        'Account': 'Shipping/Handling/Vacuum/Import Fee/etc',
        'Split': '',
        'Amount': group['Amount'].values[0],
        'Customer': '',
        'Vendor': '',
        'Class': '',
        'Qty': '',
        'Rate': '',
        'Debit': '',
        'Credit': '',
        'Business': '',
        'Product/Service': 'Shipping/Handling/Vacuum/Import Fee/etc'
    }

    # convert the dictionary to a Series and append it to the new_df DataFrame
    new_row_series = pd.Series(new_row_dict)
    new_df = new_df.append(new_row_series, ignore_index=True)
```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df['Month'] = new_df['Date'].dt.month
new_df
```

```python
new_df['Amount'].sum()
```

```python
# create a new DataFrame to store the summarized information
final_df = pd.DataFrame(columns=empty_class.columns)

for i in (ponderado['Month'].unique()):
    for date in new_df[new_df['Month']==i]['Date']:
        for porcentaje in ponderado[ponderado['Month']==i]['%']:
                    
            # create a dictionary for the new row
            new_row_dict = {
                'Date': date,
                'Transaction Type': '',
                'Num': '',
                'Name': '',
                'Memo/Description': 'Shipping/Handling/Vacuum/Import Fee/etc',
                'Account': 'Shipping/Handling/Vacuum/Import Fee/etc',
                'Split': '',
                'Amount': float(new_df[new_df['Date']==date]['Amount'] * (porcentaje / 100)),
                'Customer': '',
                'Vendor': '',
                'Class': ponderado[ponderado['%']==porcentaje]['Class'].iloc[0],
                'Qty': '',
                'Rate': '',
                'Debit': '',
                'Credit': '',
                'Business': '',
                'Product/Service': 'Shipping/Handling/Vacuum/Import Fee/etc'
            }

            # convert the dictionary to a Series and append it to the new_df DataFrame
            new_row_series = pd.Series(new_row_dict)
            final_df = final_df.append(new_row_series, ignore_index=True)
```
